```python
import requests
import pandas as pd
import json
```

```python
headers = {'X-Api-Key': 'a8cc3702-5b24-46a2-be97-75c4c7c0dcf4'}
```

```python
def run_query(query):
    request = requests.get('https://openstates.org/graphql', json={'query': query}, headers=headers)
    if request.status_code == 200:
        return request.json()
    else:
        raise Exception("Query failed to run by returning code of {}. {}".format(request.status_code, query))
```

```python
query = """
{
  people(memberOf:"ocd-organization/7e3c07c9-91ff-4445-8cef-57f85ab6ae9f", first: 100) {
    edges {
      node {
        name
        party: currentMemberships(classification:"party") {
          organization {
            name
          }
        }
        contactDetails {
            type
            value
            note
            label
        }
        links {
          url
        }
        sources {
          url
        }
        chamber: currentMemberships(classification:["upper", "lower"]) {
          post {
            label
          }
          organization {
            name
            classification
            parent {
              name
            }
          }
        }
      }
    }
  }
}
"""
```

```python
json_data = run_query(query)
```
